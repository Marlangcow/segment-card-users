```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

# 시각화 스타일 설정
plt.style.use('seaborn')
sns.set_palette('Set2')

# 경로 설정
sys.path.append('../..')
from src.data.data_loader import CreditCardDataLoader
```

```python
# 데이터 로드
data_path = '../../open/train'
loader = CreditCardDataLoader(data_path)
data = loader.load_all_data()
```

```python
# 데이터 기본 정보 확인
info = loader.get_data_info()

for name, info_dict in info.items():
    print(f"\n=== {name} 데이터 정보 ===")
    print(f"Shape: {info_dict['shape']}")
    print("\nColumns:")
    print(info_dict['columns'])
    print("\nMissing Values:")
    missing = {k: v for k, v in info_dict['missing_values'].items() if v > 0}
    if missing:
        print(missing)
    else:
        print("No missing values")
```

```python
member_df = data['member']

# 기본 통계량
print("=== 회원 데이터 기본 통계 ===")
print(member_df.describe())

# 회원 수 확인
print(f"\n총 회원 수: {len(member_df)}")
```

```python
credit_df = data['credit']

# 기본 통계량
print("=== 신용 데이터 기본 통계 ===")
print(credit_df.describe())

# 신용 점수 분포 시각화
plt.figure(figsize=(12, 6))
for col in credit_df.columns:
    if '신한최대한도' in col:
        sns.histplot(data=credit_df, x=col, bins=50)
        plt.title(f'{col} 분포')
        plt.show()
```

```python
transaction_df = data['transaction']

# 기본 통계량
print("=== 승인매출 데이터 기본 통계 ===")
print(transaction_df.describe())

# 승인매출 금액 분포
plt.figure(figsize=(12, 6))
for col in transaction_df.columns:
    if '이용금액' in col:
        sns.boxplot(data=transaction_df, y=col)
        plt.title(f'{col} 분포')
        plt.show()
```
